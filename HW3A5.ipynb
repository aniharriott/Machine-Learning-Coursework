```python
import torch
from torch.nn.functional import relu, cross_entropy
from mnist import MNIST
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# Neural Networks
mndata = MNIST('./data/')
X_train, labels_train = map(torch.tensor, mndata.load_training()) 
X_test, labels_test = map(torch.tensor, mndata.load_testing()) 
X_train = X_train/255.0
X_test = X_test/255.0

def weight_init(output_dim, input_dim):
    alpha = 1/math.sqrt(input_dim)
    weights = torch.Tensor(output_dim, input_dim).uniform_(-alpha, alpha).requires_grad_()
    bias = torch.Tensor(output_dim).uniform_(-alpha, alpha).requires_grad_()
    return weights, bias
```

```python
# A5
# part a
h = 64 # dimension of hidden layers
n, d = X_train.shape
k = 10 # number of target classes
W_0, b_0 = weight_init(h, d)
W_1, b_1 = weight_init(k, h)
parameters = [W_0, b_0, W_1, b_1]
optim_Adam = torch.optim.Adam(parameters) # more advanced gradient descent
accuracy = 0
loss_plot = [] # loss per epoch

while accuracy < 0.99:
    optim_Adam.zero_grad()
    layer_0 = X_train.matmul(W_0.t()) + b_0
    F_1 = relu(layer_0).matmul(W_1.t()) + b_1
    loss = cross_entropy(F_1, labels_train)
    loss_plot.append(loss)
    # updating parameters
    loss.backward()
    optim_Adam.step()
    accuracy = torch.mean((torch.argmax(F_1, dim=1) == labels_train).double())
    print(accuracy)
```

```python
plt.figure(figsize=(20, 10))
plt.title("Training plot, h=64")
plt.xlabel('epoch')
plt.ylabel('loss')
plt.plot(loss_plot)

# Evaluate on test set
layer_0 = X_test.matmul(W_0.t()) + b_0
F_1 = relu(layer_0).matmul(W_1.t()) + b_1
loss = cross_entropy(F_1, labels_test)
accuracy = torch.mean((torch.argmax(F_1, dim=1) == labels_test).double())
print('Loss: ', loss)
print('Accuracy:', accuracy)
```

```python
# part b
h_0 = 32 # dimension of hidden layers
h_1 = 32
n, d = X_train.shape
k = 10 # number of target classes
W_0, b_0 = weight_init(h_0, d)
W_1, b_1 = weight_init(h_1, h_0)
W_2, b_2 = weight_init(k, h_1)
parameters = [W_0, b_0, W_1, b_1, W_2, b_2]
optim_Adam = torch.optim.Adam(parameters) # more advanced gradient descent
accuracy = 0
loss_plot = [] # loss per epoch

while accuracy < 0.99:
    optim_Adam.zero_grad()
    layer_0 = X_train.matmul(W_0.t()) + b_0
    layer_1 = relu(layer_0).matmul(W_1.t()) + b_1
    F_2 = relu(layer_1).matmul(W_2.t()) + b_2
    loss = cross_entropy(F_2, labels_train)
    loss_plot.append(loss)
    # updating parameters
    loss.backward()
    optim_Adam.step()
    accuracy = torch.mean((torch.argmax(F_2, dim=1) == labels_train).double())
    print(accuracy)
```

```python
plt.figure(figsize=(20, 10))
plt.title("Training plot, h_0, h_1 = 32")
plt.xlabel('epoch')
plt.ylabel('loss')
plt.plot(loss_plot)

# Evaluate on test set
layer_0 = X_train.matmul(W_0.t()) + b_0
layer_1 = relu(layer_0).matmul(W_1.t()) + b_1
F_2 = relu(layer_1).matmul(W_2.t()) + b_2
loss = cross_entropy(F_2, labels_train)
accuracy = torch.mean((torch.argmax(F_2, dim=1) == labels_train).double())
print('Loss: ', loss)
print('Accuracy:', accuracy)
```
